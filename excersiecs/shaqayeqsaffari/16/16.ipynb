**(A AND B) OR (B OR C) XOR (A NAND B)**

```python
def Tree1(a, b, c):
  if a == True:
    if b == True:
      if c == True:
          return True 
      elif c == False:
          return False 
    elif b == False: 
      return False
  elif a == False:
    if b == False:
      if c == True:
          return False
      elif c == False: 
          return True
    elif b == True:
      if c == True:
          return True 
      elif c == False:
          return False



print("A", "B", "C", "Tree1")      
print("0", "0", "0", Tree1(0,0,0))
print("0", "0", "1", Tree1(0,0,1))
print("0", "1", "0", Tree1(0,1,0))
print("0", "1", "1", Tree1(0,1,1))
print("1", "0", "0", Tree1(1,0,0))
print("1", "0", "1", Tree1(1,0,1))
print("1", "1", "0", Tree1(1,1,0))
print("1", "1", "1", Tree1(1,1,1))
```

```python
import pandas
from sklearn import tree
import pydotplus
from sklearn.tree import DecisionTreeClassifier
import matplotlib.pyplot as plt
import matplotlib.image as pltimg


df = pandas.read_csv("/content/concept1")

print(df)
print("______________________")

features = ['A', 'B', 'C']

X = df[features]
y = df['concept']


dtree = DecisionTreeClassifier()
dtree = dtree.fit(X, y)
data = tree.export_graphviz(dtree, out_file=None, feature_names=features)
graph = pydotplus.graph_from_dot_data(data)
graph.write_png('concept.jpeg')

img=pltimg.imread('concept.jpeg')
imgplot = plt.imshow(img)
plt.show()
```

**(A XOR B) AND (B OR C) AND (C AND D)**

```python
def Tree3(a, b, c, d):
    if a == True:
        if b == True:
              return True 
        elif b == False:
              if c == False:
                  return False 
              elif c == True: 
                  if d == False:
                        return False
                  elif d == True:
                        return True
    elif a == False:
        if b == False:
              return False 
        elif b == True:
              if c == False:
                  return False 
              elif c == True:
                  if d == False:
                        return False
                  elif d == True:
                        return True



print("A", "B", "C", "D", "Tree3")      
print("0", "0", "0", "0", Tree3(0,0,0,0))
print("0", "0", "0", "1", Tree3(0,0,0,1))
print("0", "0", "1", "0", Tree3(0,0,1,0))
print("0", "0", "1", "1", Tree3(0,0,1,1))
print("0", "1", "0", "0", Tree3(0,1,0,0))
print("0", "1", "0", "1", Tree3(0,1,0,1))
print("0", "1", "1", "0", Tree3(0,1,1,0))
print("0", "1", "1", "1", Tree3(0,1,1,1))
print("1", "0", "0", "0", Tree3(1,0,0,0))
print("1", "0", "0", "1", Tree3(1,0,0,1))
print("1", "0", "1", "0", Tree3(1,0,1,0))
print("1", "0", "1", "1", Tree3(1,0,1,1))
#            
```
